```python
import random
import google.generativeai as genai
from google.colab import userdata
import subprocess

def verificar_pacote_instalado(pacote):
    try:
        resultado = subprocess.check_output(["pip", "show", pacote])
        return True
    except subprocess.CalledProcessError:
        return False

def instalar_pacote(pacote):
    subprocess.call(["pip", "install", "-qU", pacote])

pacote_desejado = "google-generativeai"
if not verificar_pacote_instalado(pacote_desejado):
    instalar_pacote(pacote_desejado)

GOOGLE_API_KEY=userdata.get("SECRET_KEY")
genai.configure(api_key=GOOGLE_API_KEY)
genaration_config = {
    "candidate_count": 1,
    "temperature": 0.5,
}
safety_settings = {
    "HARASSMENT": "BLOCK_NONE",
    "HATE": "BLOCK_NONE",
    "SEXUAL": "BLOCK_NONE",
    "DANGEROUS": "BLOCK_NONE",
}
model = genai.GenerativeModel(
    model_name="gemini-1.0-pro",
    generation_config=genaration_config,
    safety_settings=safety_settings
    )

chat = model.start_chat(history=[])

vNome = input("Olá, qual é o seu nome? ")
if vNome == "":
  vNome = "Estudante"

print(f"\nPrazer em te conhecer, {vNome}! Vamos aprender a tabuada juntos? \n")

while True:
    tabuada_escolhida = input("Qual tabuada você quer praticar? (1 a 10 ou 0 para todas): ")
    if len(tabuada_escolhida) <= 2 and (tabuada_escolhida.isdigit() or tabuada_escolhida == '.') and (tabuada_escolhida == '.' or 0 <= float(tabuada_escolhida) <= 10):
        break
    else:
        print("Por favor, digite um número de 0 a 10 ou um ponto (.) para finalizar.")

if tabuada_escolhida.lower() == ".":
  print(f"Entendo {vNome}, sem problemas. Espero te ver novamente.\n\n---\n")
  response = chat.send_message("Mensagem curta e motivacional sobre estudar matemática.")
  print(f"{response.text}\n")
else:
  if tabuada_escolhida.lower() == str(0):
    tabuada_da_vez = random.randint(1, 10)
    print(f"\nLegal, vamos estudar começando pela tabuada do {tabuada_da_vez}. \n\n")
  else:
    tabuada_da_vez = tabuada_escolhida
    print(f"\nLegal, vamos estudar a tabuada do {tabuada_da_vez}. \n\n")

  response = chat.send_message("Conte uma curiosidade sobre a tabuada do "+str(tabuada_da_vez)+".")
  print(f"Curiosidade: {response.text} \n\n")
  pergunta_crianca = ""
  print("Pronto! Vamos iniciar nosso estudo. Digite ponto (.) quando quiser terminar. \n")

  while pergunta_crianca != ".":
    numero = random.randint(1, 10)
    resultado = numero * int(tabuada_da_vez)

    while True:
        pergunta_crianca = input(f"{vNome}, quanto é {numero} x {tabuada_da_vez}? ")
        if (tabuada_escolhida.isdigit() or tabuada_escolhida == '.'):
            break
        else:
            print("Por favor, digite apenas números ou um ponto (.) para finalizar.")

    if pergunta_crianca.lower() == ".":
      break
    elif pergunta_crianca.lower() == str(resultado):
        response = chat.send_message("Escreva uma mensagem curta e única de parabenização para uma criaça que acertou a tabuala do "+str(tabuada_da_vez)+" onde "+str(numero)+" x "+str(tabuada_da_vez)+" = "+str(resultado)+". O nome da criança é"+vNome+".")
        print(response.text, "\n")
    else:
        print(f"Que pena {vNome}. A resposta correta é {resultado}. Continue tentando!")

    if tabuada_escolhida.lower() == str(0):
      tabuada_da_vez = random.randint(1, 10)

  print(f"\nObrigado por estudar comigo, {vNome}! Você aprendeu muito hoje!")
  print(f"\nEspero nos vermos em breve.\n---\n")
  response = chat.send_message("Escreva um pensamento de algum matemático famoso.")
  print(response.text, "\n---\n")

pergunta_conversar = input("Agora que seus estudos foram concluídos, você gostaria de conversar sobre outro assunto? (Sim/Não): ")
if pergunta_conversar.lower() == "sim" or pergunta_conversar.lower() == "s":
  prompt = ""
  print("Que bacana, é muito bom conversar com você. Quando conluir, digite fim para sair. \n")
  while prompt != "fim":
    prompt = input(vNome+": ")
    response = chat.send_message(prompt)
    print("Gemini: ", response.text, "\n")

  print(f"\n---\nAté mais {vNome}. Espero conversar mais com você em breve.")
else:
  print("Sem problema, até mais.")
```
